# Data Analysis & Results

```python
#import package for statistical tests
from scipy import stats
import statsmodels.formula.api as sfa
from statsmodels.formula.api import ols
import statsmodels.api as sm
import sys
!{sys.executable} -m pip install scikit_posthocs
import scikit_posthocs as sp
```

First we create a simple function that compares the test p-value to the alpha level which will allow us to either reject or fail to reject the null hypothesis.

```python
#define the alpha level
significance = 0.05

#Create the function
def check_pval(pvalue):
    if pvalue > significance:
        return 'Fail to reject the null hypothesis'
    if pvalue < significance:
        return 'Reject the null hypothesis'
```

In order to test our first hypothesis as to whether there are discrepancies between ethnic groups’ access to telemedicine in the U.S., we will be using an ANOVA to compare the means across ethnicity groups within each early and late time period. 

First we'll produce a boxplot to see the distribution of each ethnicity within the early time period.

```python
ax = sns.boxplot(x='ethnicity', y='case_count', data=early_melt)
#ax = sns.swarmplot(x='ethnicity', y='case_count', data=early_melt, color='#7d0013')
```

Now we conduct a one-way ANOVA to compare the mean across all ethnicity groups regarding the data in the early_infection dataframe.

```python
#ANOVA between ethnicity group data during the early infection timepoint
fvalue_early, pvalue_early = stats.f_oneway(early_infection['White'], early_infection['Hispanic'],
                               early_infection['Black Non-Hispanic'], early_infection['Other Non-Hispanic'])
print(fvalue_early, pvalue_early)

#Apply the check function to the computed p-value
print(check_pval(pvalue_early))

#Ordinary Least Squares model
model_early = ols('case_count ~ C(ethnicity)', data=early_melt).fit()
anova_early = sm.stats.anova_lm(model_early, typ=2)
print(anova_early)
```

The result of this test indicates that our p-value is less than the alpha level (0.000074 < 0.05). Therefore, we can reject the null hypothesis and conclude that there are significant differences in the mean cases of access to telemedicine across all ethnicity groups during the early time period.

Now that we know that the ethnicity groups are not statstically equal, we must perform a multiple pairwise comparison to see which specific ethnicity groups are statistically different. We'll be specifically using the Tukey HSD test.

```python
early_posthoc = sp.posthoc_ttest(early_melt, val_col='case_count', group_col='ethnicity', p_adjust='holm')
print(early_posthoc)

print('\nWhite comparison')
print(early_posthoc['White'].apply(check_pval))

print('\nHispanic comparison')
print(early_posthoc['Hispanic'].apply(check_pval))
print()

print('\nBlack Non-Hispanic comparison')
print(early_posthoc['Black Non-Hispanic'].apply(check_pval))
print()

print('\nOther Non-Hispanic comparison')
print(early_posthoc['Other Non-Hispanic'].apply(check_pval))
print()
```

With these results, our p-values that are less than the alpha level 0.05 indicates that we can reject the null hypothesis that there are no significant differences amongst ethnicity groups. In particular the following groups indicate a statstical difference in the mean cases: White and Black Non-Hispanic, White and Hispanic, Hispanic and Other Non-Hispanic, and Black Non-Hispanic and Other Non-Hispanic.
  
The following groups have a p-value greater than the alpha level 0.05 and therefore we preserve the null hypothtesis that these ethnicity groups do not indicate a statistical difference in mean cases: White and Other Non-Hispanic, and Black Non-Hispanic and Hispanic.

Now we'll produce a boxplot to see the distribution of each ethnicity within the late time period.

```python
ax = sns.boxplot(x='ethnicity', y='case_count', data=late_melt)
```

Now we apply a one-way ANOVA to compare the mean across all ethnicity groups regarding the data in the late_infection dataframe.

```python
#ANOVA between ethnicity group data during the early infection timepoint
fvalue_late, pvalue_late = stats.f_oneway(late_infection['White'], late_infection['Hispanic'],
                               late_infection['Black Non-Hispanic'], late_infection['Other Non-Hispanic'])
print(fvalue_late, pvalue_late)

#Apply the function to the p-value
check_pval(pvalue_late)

#Ordinary Least Squares model
model_late = ols('case_count ~ C(ethnicity)', data=late_melt).fit()
anova_late = sm.stats.anova_lm(model_late, typ=2)
print(anova_late)
```

The result of this test indicates that our p-value is less than the alpha level (0.0002 < 0.05). Therefore, we can reject the null hypothesis and conclude that there are significant differences in the mean cases of access to telemedicine across all ethnicity groups during the late time period.

Now that we know that the ethnicity groups are not statstically equal, we must again perform a multiple pairwise comparison to see which specific ethnicity groups are statistically different. We'll be specifically using the Tukey HSD test.

```python
late_posthoc = sp.posthoc_ttest(late_melt, val_col='case_count', group_col='ethnicity', p_adjust='holm')
print(late_posthoc)

print('\nWhite comparison')
print(late_posthoc['White'].apply(check_pval))

print('\nHispanic comparison')
print(late_posthoc['Hispanic'].apply(check_pval))
print()

print('\nBlack Non-Hispanic comparison')
print(late_posthoc['Black Non-Hispanic'].apply(check_pval))
print()

print('\nOther Non-Hispanic comparison')
print(late_posthoc['Other Non-Hispanic'].apply(check_pval))
print()
```

With these results, our p-values that are less than the alpha level 0.05 indicates that we can reject the null hypothesis that there are no significant differences amongst ethnicity groups. In particular the following groups indicate a statstical difference in the mean cases: White and Black Non-Hispanic, and Black Non-Hispanic and Other Non-Hispanic.
  
The following groups have a p-value greater than the alpha level 0.05 and therefore we preserve the null hypothtesis that these ethnicity groups do not indicate a statistical difference in mean cases: White and Hispanic, White and Other Non-Hispanic, Black and Hispanic, and Hispanic and Other Non-Hispanic.

In order to test the second hypothesis we must use a independent sample t-test to compare each ethnicity's mean value in the early_infection and late_infection dataframe.

```python
#Independent t-test for White ethnicity across early and late time period
white_comp = stats.ttest_ind(early_infection['White'], late_infection['White'])
white_pval = white_comp.pvalue

#Independent t-test for Hispanic ethnicity across early and late time period
hispanic_comp = stats.ttest_ind(early_infection['Hispanic'], late_infection['Hispanic'])
hispanic_pval = hispanic_comp.pvalue

#Independent t-test for Black Non-Hispanic ethnicity across early and late time period
black_comp = stats.ttest_ind(early_infection['Black Non-Hispanic'], late_infection['Black Non-Hispanic'])
black_pval = black_comp.pvalue

#Independent t-test for Other Non-Hispanic ethnicity across early and late time period
other_comp = stats.ttest_ind(early_infection['Other Non-Hispanic'], late_infection['Other Non-Hispanic'])
other_pval = other_comp.pvalue

#Check values
print(white_comp)
print(check_pval(white_pval))
print()

print(hispanic_comp)
print(check_pval(hispanic_pval))
print()

print(black_comp)
print(check_pval(black_pval))
print()

print(other_comp)
print(check_pval(other_pval))
print()
```

With the given data, we are able to reject the null hypothesis because the p-value across all early and late ethnicity comparisons is less than the alpha level of 0.05; therefore we can conclude that the usage of telemedicine increased overall across all ethnicity groups from the beginning of the COVID-19 pandemic to now.

# Conclusion & Discussion

$\;\;\;\;\;\;$Through our statistical analysis we are able to conclude that there are in fact statistical differences in the mean total of telemedicine cases amongst ethnicity groups in both the early and late time period. Our hypothesis had predicted that these discernable differences were possibly due to social conditions surrounding race, therefore we had anticipated a result of White telemedicine cases significantly differing from that of ethnic minorities. 

$\;\;\;\;\;\;$In the early dataset we were able to reject the null hypothesis and conclude that pairs of ethnicity groups were statistically different within the following comparisons: White and Black Non- Hispanic, White and Hispanic, Hispanic and Other Non-Hispanic, and Black Non-Hispanic and Other Non-Hispanic. For our conclusions we will factor out the category of Other Non-Hispanic individuals due to it’s vague categorization and factoring of many different races, which skews its group numbers and accounts for 38.4% of early telemedicine cases. With this, we can see that White cases (p=30.1%) when compared to Black cases (p=18.0%) and Hispanic cases (p=13.0%) make up nearly twice as much of the dataset’s telemedicine usage. This may be due to factors such as ethnic minorities having less overall access to health care services due to higher uninsurance rates, language and cultural barriers, or technological barriers. However, we see different results when looking at a later, more recent time period.

$\;\;\;\;\;\;$In the late dataset we were also able to reject the null hypothesis and conclude that pairs of ethnicity groups were statistically different within the following comparisons: White and Black Non-Hispanic, and Black Non-Hispanic and Other Non-Hispanic. Again we exclude the data from the Other Non-Hisapnic category for its skewed numbers, which allows it to make up 59.4% of the telemedicine cases. With this we can see that the only statistically different comparisons are between that of White cases (p=40.0%) and Black cases (p=9.0%). Some explanations for the shift in differences among ethnicity groups may be explained by the overall increase in telemedicine cases from the early time period to the late time period. The late_infection dataframe has significantly more telemedicine counts than the early_infection dataframe (19,296,572 > 374,279) which is reasonable as many health care providers switched to a minimal contact delivery of medical services such as telemedicine. The increase of telemedicine usage can also be explained through the expansion of Medicare telehealth during the COVID-19 public health emergency. With this, previously uninsured individuals, including ethnic minorities who may be disproportionately uninsured, could have gained health coverage when compared to the earlier time point. As previously mentioned, we did analyze an overall increase in telemedicine usage across all ethnicity groups, through a series of t-tests which can also be explained by the shift to telemedicine as the primary delivery of healthcare and expansions of government aid.

$\;\;\;\;\;\;$Overall our data did allow us to answer our research question as to whether there were discrepancies between ethnicity groups and whether within each group total cases of telemedicine had increased or decreased during the course of the COVID-19 pandemic. However, our dataset is limited in the conclusions we can make. Although we speculate possible social conditions surrounding race as a factor affecting differences amongst ethnicity groups, we do not have operational definitions to measure and further validate these possible reasons. For example, our datasets did not contain information regarding socioeconomic status such as average household income for ethnicity groups within each state. With this data we would be able to compare these numbers and be able to factor SES into why some ethnicity groups may be more likely to be uninsured than others (e.g. Lesser job stability and therefore no employer-sponsored insurance; inability to cover health service premiums; etc.). Another limitation is that the Other Non-Hispanic categorization factored in many races and skewed the data in comparison to the other ethnicity categories. Races such as Asian, American Indian and Alaska Native, and Native Hawaiian and Pacific Islander vary significantly in predisposed health conditions, employment status and SES and potential social barriers which may affect their access to healthcare.

$\;\;\;\;\;\;$In conclusion, access to telemedicine is a critically important resource during the circumstances of a pandemic, in which making frequent trips to a clinic in search of medical care can prove risky, especially to the immunocompromised and elderly. In addition, different ethnicity groups in the United States have historically received unequal access to healthcare depending on a variety of factors. Our project set to examine whether potential barriers existed between different ethnic groups within the specific framework of telemedicine access during COVID-19. Through our research, we discovered that, in general, participants belonging to historically-disadvantaged ethnic groups sought out and received telemedicine care at significantly different percentages compared to white participants, but the circumstances of this discrepancy changed throughout the course of the pandemic. We may speculate a variety of reasons for why this is; perhaps the expanding of social service programs, increase of total case numbers, differences in infection rates between different ethnic communities all play a factor of some degree. However, what we can clearly identify is that there exists observable imbalances in the overall amount of telemedicine demand/usage that can be studied through the total participants of each ethnic group that actually accessed such services during the course of the pandemic. This points to clear inequalities between standards and needs of care between these groups, and is an important set of statistics to examine for the purpose of establishing, diagnosing, and addressing larger disparities in access to healthcare, social infrastructure, and government services both within the context of a pandemic, as well as outside of one. These are issues that have existed prior to the COVID-19 pandemic, continue to exist now, and may well extend into the future without proper examination of the true extent of their prevalence.
