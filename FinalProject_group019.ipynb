# Background and Prior Work

### Why is this question important to our group?


Our group wanted to research the topic of rape as it is a very common and large issue in the United States. ### What do we already know?

The U.S. Department of Justice defines rape as the “penetration, no matter how slight, of the vagina or anus with any body part or object, or oral panetration by a sex organ of another person, without the consent of the victim” (3). Rape incidences affects more than just the victim, it hinders familial and work relationships, as well as the entire community. According to the National Sexual Violence Resource Center, the nationcal economic cost of rape is  $127 billion per year which is more than any other crime (2). Unfortunately, the current rate of victims of rape is expected to continue to increase over the years. Through research we have found that lower education levels, the use of drugs and alcohol, and financial hardships are all contributors to rape and other forms of sexual assault (4). We believe that we can validate these claims by investigating our data. We predict that impoverished areas have a larger number of rape cases than those of affluent areas.

### Normalize and rename Drug 2018 

```python
#rename column
drug2018.rename(columns = {"Illicit Drug Use in percentage for year 12+": "Drug Use"}, inplace = True)

#normalize
column = drug2018["Drug Use"]
Max = column.max()
for d in drug2018["Drug Use"]:
    drug2018["Drug Use"].replace({d : d / Max}, inplace = True)

#rename state column
for d in drug2018["state"]:
    st = d + str(" 2018")
    drug2018["state"].replace({d : st}, inplace = True)

#rename column
drug2018.rename(columns = {"state": "State"}, inplace = True)
#drug2018
```

### Normalize and rename Drug 2019 column as well as fix certain values

```python
#rename column
drug2019.rename(columns = {"Drug Use Score": "Drug Use"}, inplace = True)

#fix certain columns
drug2019["Drug Use"].replace({58396.00 : 58.96}, inplace = True)
drug2019["Drug Use"].replace({5832.00 : 58.32}, inplace = True)
drug2019["Drug Use"].replace({429.00 : 42.90}, inplace = True)

#normalize
column = drug2019["Drug Use"]
Max = column.max()
for d in drug2019["Drug Use"]:
    drug2019["Drug Use"].replace({d : d / Max}, inplace = True)

#rename state column
for d in drug2019["State"]:
    st = d + str(" 2019")
    drug2019["State"].replace({d : st}, inplace = True)

#sort State alphabetically
drug2019 = drug2019.sort_values(by='State', ascending = True)
#drug2019
```

### Normalize and rename / delete certain columns for Education 2018

```python
#add average of 2 existing columns as new columns to dataframe
edu = []
for i in range(len(education2018)):
    perc = (education2018.loc[i, "PercentHighSchoolOrHigher"] + education2018.loc[i, "PercentBachelorsOrHigher"]) / 2
    edu.append(perc)
education2018['Education Rank'] = edu

#normalize
column = education2018["Education Rank"]
Max = column.max()
for d in education2018["Education Rank"]:
    education2018["Education Rank"].replace({d : d / Max}, inplace = True)

#remove two old columns
del education2018['PercentHighSchoolOrHigher']
del education2018['PercentBachelorsOrHigher']

#add year to state name
for d in education2018["State"]:
    st = d + str(" 2018")
    education2018["State"].replace({d : st}, inplace = True)

education2018 = education2018.sort_values(by='State', ascending = True)
#education2018
```

### Edit / Normalize Education 2019 by taking the average of all 5 original columns then dividing by max. Then rename column.

```python
education2019.loc[0]

#avg of all 5 categories / add to dataframe as separate column
edu = []
for i in range(len(education2019)):
    x1 = education2019.loc[i, "Academic Performance Rank"]
    x2 = education2019.loc[i, "Safety Rank"]
    x3 = education2019.loc[i, "Community Rank"]
    x4 = education2019.loc[i, "Academic Investment Rank"]
    x5 = education2019.loc[i, "Class Size and Attendance Rank"]
    avg = (x1 + x2 + x3 + x4 + x5)  / 5
    edu.append(avg)
education2019["Education Rank"] = edu

#normalize
column = education2019["Education Rank"]
Max = column.max()
for d in education2019["Education Rank"]:
    education2019["Education Rank"].replace({d : d / Max}, inplace = True)

#remove five old columns
del education2019["Academic Performance Rank"]
del education2019["Safety Rank"]
del education2019["Community Rank"]
del education2019["Academic Investment Rank"]
del education2019["Class Size and Attendance Rank"]

#add year to state name
for d in education2019["State"]:
    st = d + str(" 2019")
    education2019["State"].replace({d : st}, inplace = True)

education2019 = education2019.sort_values(by='State', ascending = True)
#education2019
```

### Normalize Income 2018 - 2020 and add years to state names

```python
#income2018
column = income2018["Median Household Income"]
Max = column.max()
for d in income2018["Median Household Income"]:
    income2018["Median Household Income"].replace({d : d / Max}, inplace = True)

for d in income2018["State"]:
    st = d + str(" 2018")
    income2018["State"].replace({d : st}, inplace = True)

income2018 = income2018.sort_values(by='State', ascending = True)

#income2019
column = income2019["Median Household Income"]
Max = column.max()
for d in income2019["Median Household Income"]:
    income2019["Median Household Income"].replace({d : d / Max}, inplace = True)

for d in income2019["State"]:
    st = d + str(" 2019")
    income2019["State"].replace({d : st}, inplace = True)

income2019 = income2019.sort_values(by='State', ascending = True)

#income2020
income2020.rename(columns = {"Average Household Income": "Median Household Income"}, inplace = True)
column = income2020["Median Household Income"]
Max = column.max()
for d in income2020["Median Household Income"]:
    income2020["Median Household Income"].replace({d : d / Max}, inplace = True)

for d in income2020["State"]:
    st = d + str(" 2020")
    income2020["State"].replace({d : st}, inplace = True)

income2020 = income2020.sort_values(by='State', ascending = True)
```

### Normalize rape cases by state by dividing by state population of given year

```python
#rape cases 2018-2019

#Sort by State
rape2018 = rape2018.sort_values(by='State', ascending = True)
rape2019 = rape2019.sort_values(by='State', ascending = True)

#2018
rapeCases = []
for i in range(len(rape2018)):
    state = rape2018.loc[i, "State"]
    cases = rape2018.loc[i, "totalRapes"]
    pop = pop2018.loc[i, "2018 Population"]
    perc = cases / pop
    rapeCases.append(perc)

rape2018['Rape Cases'] = rapeCases
del rape2018['totalRapes']

#normalize
column = rape2018['Rape Cases']
Max = column.max()
for d in rape2018['Rape Cases']:
    rape2018['Rape Cases'].replace({d : d / Max}, inplace = True)

#add year to state name
for d in rape2018["State"]:
    st = d + str(" 2018")
    rape2018["State"].replace({d : st}, inplace = True)

#2019
rapeCases = []
for i in range(len(rape2019)):
    state = rape2019.loc[i, "State"]
    cases = rape2019.loc[i, "Sexual Assault Cases "]
    pop = pop2019.loc[i, "POPESTIMATE2019"]
    perc = cases / pop
    rapeCases.append(perc)

rape2019['Rape Cases'] = rapeCases
del rape2019['Sexual Assault Cases ']

#normalize
column = rape2019['Rape Cases']
Max = column.max()
for d in rape2019['Rape Cases']:
    rape2019['Rape Cases'].replace({d : d / Max}, inplace = True)

#add year to state name
for d in rape2019["State"]:
    st = d + str(" 2019")
    rape2019["State"].replace({d : st}, inplace = True)
```

#### Linear Regression

```python
df.rename(columns = {'Drug Abuse':'Drug'}, inplace = True)
outcome, predictors = patsy.dmatrices('Rape ~ Drug', df)
model = sm.OLS(outcome, predictors)
```

```python
results = model.fit()
print(results.summary())
```

The r-squared in this case is 0.0000, which means the r value will be around 0 as well. Therefore, there is an absence of linear relationship between drug abuse and rape rates. Based on the facts, we fail to reject the null hypothesis which drug abuse has no effect on rape.

#### Plot Best-Fit Line

```python
sns.scatterplot(x='Drug', y='Rape', data=df, s=50).set_title('Drug v Rape')

# Generate and plot the model fit line
a1, b1 = np.polyfit(df['Drug'].values, df['Rape'].values, 1)
yfit = [b1 + a1 * xi for xi in df['Drug']]
plt.plot(df['Drug'], yfit, 'Red')

plt.xlabel('Drug')
plt.ylabel('Rape')
```

When analyzing the linear relationship between Drug Use and Rape, the best fit line has a slope of almost 0 implying that Drug does not change the intercept value of Rape.

Our project will only be investigating the US population, however, sampling bias still takes place since many datasets we used are collected through surveys.
