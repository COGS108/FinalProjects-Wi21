# Data Analysis & Results

```python
df_unemployment = pd.DataFrame(df_occupt, columns =['Occupation','Unemployed (Dec. 2019)','Unemployed (Dec. 2020)', 'Unemployment Rate (Dec. 2019)','Unemployment Rate (Dec. 2020)'])

# Data cleaning by changing the string to floats
df_unemployment['Unemployed (Dec. 2019)'] = df_unemployment['Unemployed (Dec. 2019)'].str.replace(',', '').astype(float)
df_unemployment['Unemployed (Dec. 2020)'] = df_unemployment['Unemployed (Dec. 2020)'].str.replace(',', '').astype(float)
# Data cleaning by removing the word 'occupations' from the columns of the dataset
df_unemployment['Occupation' ] = df_unemployment['Occupation'].str.replace(' occupations', '')

# Create a bar chart to visualize Number of People Unemployed in Dec. 2019 by Occupation
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(df_unemployment['Occupation'].unique(), df_unemployment['Unemployed (Dec. 2019)'])
plt.xticks(rotation = 90)
plt.ylabel('Number of People')
plt.xlabel('Occupation')
plt.title('Number of People Unemployed in Dec.2019 by Occupation')

# Create a bar chart to visualize Unemployment Rate in Dec. 2019 by Occupation
plt.subplot(1,2,2)
plt.bar(df_unemployment['Occupation'].unique(), df_unemployment['Unemployment Rate (Dec. 2019)'])
plt.xticks(rotation = 90)
plt.ylabel('Unemployment Rate')
plt.xlabel('Occupation')
plt.title('Unemployment Rate in Dec. 2019 by Occupation')
plt.show()

# Create a bar chart to visualize Number of People Unemployed in Dec. 2020 by Occupation
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(df_unemployment['Occupation'].unique(), df_unemployment['Unemployed (Dec. 2019)'])
plt.bar(df_unemployment['Occupation'].unique(), df_unemployment['Unemployed (Dec. 2020)'])
plt.xticks(rotation = 90)
plt.ylabel('Number of People')
plt.xlabel('Occupation')
plt.title('Number of People Unemployed in Dec.2020 by Occupation')

# Create a bar chart to visualize Unemployment Rate in Dec. 2020 by Occupation
plt.subplot(1,2,2)
plt.bar(df_unemployment['Occupation'].unique(), df_unemployment['Unemployment Rate (Dec. 2019)'])
plt.bar(df_unemployment['Occupation'].unique(), df_unemployment['Unemployment Rate (Dec. 2020)'])
plt.xticks(rotation = 90)
plt.ylabel('Unemployment Rate')
plt.xlabel('Occupation')
plt.title('Unemployment Rate in Dec. 2020 by Occupation')
plt.show()


df_unemployment_rate = pd.DataFrame(df_unemployment,columns=['Occupation','Unemployment Rate (Dec. 2019)','Unemployment Rate (Dec. 2020)'])
df_unemployment_num = pd.DataFrame(df_unemployment,columns=['Occupation','Unemployed (Dec. 2019)','Unemployed (Dec. 2020)'])

fig = plt.figure() # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes
width = 0.4

# Create a bar chart to visualize Number of People Unemployed in Dec. 2019 and Dec. 2020 by Occupation
df_unemployment_num.plot(x='Occupation', y='Unemployed (Dec. 2019)', kind='bar', color='blue', ax=ax, width=width, position=1, figsize = (16,4), title = 'Number of People Unemployed in Dec. 2019 and Dec. 2020 by Occupation')
df_unemployment_num.plot(x='Occupation', y='Unemployed (Dec. 2020)', kind='bar', color='orange', ax=ax, width=width, position=0)
ax.set_ylabel('Number of People')
plt.show()

fig = plt.figure() # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes
width = 0.4

# Create a bar chart to visualize Unemployment Rate in Dec. 2019 and Dec. 2020 by Occupation
df_unemployment_rate.plot(x='Occupation', y='Unemployment Rate (Dec. 2019)', kind='bar', color='blue', ax=ax, width=width, position=1, figsize = (16,4), title = 'Unemployment Rate in Dec. 2019 and Dec. 2020 by Occupation')
df_unemployment_rate.plot(x='Occupation', y='Unemployment Rate (Dec. 2020)', kind='bar', color='orange', ax=ax, width=width, position=0)
ax.set_ylabel('Unemployment Rate')
plt.show()
```

We used descriptive analysis to summarize past data on unemployment rates in December of 2019 and in December of 2020, and compare the two using bar graphs in order to present how different occupations are affected by the COVID-19 pandemic. Then we drew an inference from the given graphs to answer our question of which careers the COVID-19 pandemic has impacted the most in the span of a year.

We can see from the last two charts that both the number of people unemployed and the unemployment rates have increased overall since Dec.2019, when the first coronavirus case was reported to the World Health Organization. Occupations which were impacted the most were those relating to service, and installation, maintenance and repair, which have nearly doubled in unemployment rates. The rate has gone from about 4 percent to about 11 percent for services, and about 2 percent to about 5 percent for installation, maintenance, and repair. The only occupation listed that didn't have much of a change in unemployment rate was within the production occupation, which slightly went up from about 4 percent to about 6 percent.

# Ethics & Privacy

First, the data that we are analyzing are reported unemployments, which means that there is a likely chance that there are unrepresented unemployments and job losses across the different careers that we are analyzing. Some job losses being reported from the months analyzed may not be a result of the COVID-19 pandemic and can be from other reasons, but are still combined as "unemployment" which could be another factor to consider. 
 
Another ethical consideration with our investigation is that the information being used is that of individuals in society, and while personal information remains confidential in the dataset, that information could still be compromised and would pose the risk of revealing personal information such as income status or ethnicity. 

The data we are analyzing is data made public from the U.S. Bureau of Labor Statistics (BLS), and the publishing of unemployment rates do not necessarily correlate with factors such as economic revenue. Furthermore, data from the BLS also may contain biases such as a nonresponse bias. According to the BLS, data is collected through "personal visits, mail, telephone, and email" which may impact whether or not an individual responds to the surveys. 

Since topics regarding employment, or lack thereof, are personal and sensitive for most, some people may choose to not respond to such surveys. This would mean that the data would be slightly hindered and may not be the most accurate sample of the whole population, however would provide for a general idea for trends with the population. As a result, training and testing the data would not necessarily be the most efficient way in analyzing the effects of unemployment for the population as a whole.

# Conclusion & Discussion

From the data we cleaned and analyzed, we found that nearly every occupation listed faced significantly higher unemployment rates from December 2020 as opposed to December 2019. Among the highest rises in unemployment were jobs in service, and installation, maintenance and repair, while production occupations were much less impacted. While we did predict that service jobs, particularly those that rely heavily on in-person interactions, would take a significant hit, we did not specifically predict the same for jobs in the latter field. However, this does make intuitive sense  installation, maintenance and repair jobs are in-person tasks, and were largely cut down on because of the pandemic. 
 
It was interesting to see that production occupations were hit the least hard due to the pandemic; this is one of the fields that we would have expected to take a hit, considering that production often entails being present in a work environment and working with other employees. Although production occupations likely dont tend to make much contact with non-employees, it is interesting that while other jobs that also stay within their bubbles cut down on employees, production occupations did not see this trend nearly as significantly. 

While we observed trends in the data that we analyzed, there are some limitations to the findings of our analysis. The data that we are analyzing is comparing unemployment in the United States in December 2019 as compared to December 2020, which means that data from other months is omitted from our analysis. Furthermore, effects of factors such as climate and peak season in retail may also play a role in changes in employment but the data being analyzed would not reflect this due to the data being from the same months (in different years). Additionally, the economic impact of the pandemic is global, yet our data only accounts for the impact on employment in the United States.
