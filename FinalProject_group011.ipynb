### Normal Test for PreCovid and Covid Stock Price

The function call normaltest in the scipy library states that the null hypothesis is that the data is normal. We notice that all the p values are significantly smaller than an alpha value of 0.05, so we reject the null hypothesis in favor of a not normally distributed dataset. As a result, we want to normalize our 'Price' data that we are plotting for both during covid pandemic and before covid pandemic for both pharmaceutical and general DJI index stocks.

From the scipy documentation* 
Null Hypothesis: The average percent change remained the same before and during H1N1.
Alternative Hypothesis: The average percent change did not remain the same before and during H1N1.

The t-test here helps us quantify and confirm the difference between 2 means. We chose this to help us indicate whether the means and standard deviations of percent change that we calculated in the above cells is significantly different. A T-score of greater than 1 indicates that there is significant difference and the greater the magnitude  the better. Here we see that both pharmaceutical and Dow Jones stocks’ percent change was significantly different from the 2 years prior to the H1N1 pandemic, thus confirming that stock prices were more volatile during the pandemic. The p-value indicates that we reject the null hypothesis in favor of the alternative hypothesis. The lower the p-value the higher the chance that this change did not arise from random chance, thus adding a layer of consistency to our results. 


```python
# perform t-test on pharma stocks for year before compared to year during h1n1
tResPharm, pResPharm = stats.ttest_ind(abs(h1n1PharmStocks['Pct Change']), abs(prevH1n1PharmStocks['Pct Change']), nan_policy = 'omit')
print('Pharmacy Stocks T-Score (significance of difference): ' + str(round(tResPharm, 3)) 
      + '\nPharmacy Stocks P-Value (probabilty this difference is from chance): ' + str(pResPharm))

# perform t-test on DJI stocks for year before compared to year pre h1n1
tResDJI, pResDJI = stats.ttest_ind(abs(h1n1DJI['Pct Change']), abs(prevH1n1DJI['Pct Change']), nan_policy = 'omit')
print('DOW Jones Stocks T-Score (significance of difference): ' + str(round(tResDJI, 3)) 
      + '\nDOW Jones Stocks P-Value (probabilty this difference is from chance): ' + str(pResDJI))
```

```python
# calculate for DJI during Sars
sarsAvgPercentChangeDJI = abs(sarsDJI['Norm Pct Change'])*100
sars_meanDJI, sars_stdDJI = st.t.interval(0.95
                                            , len(sarsAvgPercentChangeDJI)-1
                                            , loc=np.mean(sarsAvgPercentChangeDJI)
                                            , scale=st.sem(sarsAvgPercentChangeDJI))

# calculate for DJI  before Sars
presarsAvgPercentChangeDJI = abs(presarsDJI['Norm Pct Change'])*100
presars_meanDJI, presars_stdDJI = st.t.interval(0.95, len(presarsAvgPercentChangeDJI)-1, 
                                                  loc=np.mean(presarsAvgPercentChangeDJI), 
                                                  scale=st.sem(presarsAvgPercentChangeDJI))
# output
print('Average Percent Change of DOW Stocks During SARS +- Standard Deviation = ' + str(round(sars_meanDJI, 3)), "+-", str(round(sars_stdDJI, 3))
     + '\n' + 'Average Percent Change of DOW Stocks Before SARS +- Standard Deviation = ' + str(round(presars_meanDJI, 3)), "+-", str(round(presars_stdDJI, 3)))
```

Null Hypothesis: The average percent change remained the same before and during SARS.
Alternative Hypothesis: The average percent change did not remain the same before and during SARS.


Here, we calculate the t-test and p-value to compare the difference between the average percent change of pharmaceutical stocks and the Dow Jone stock before SARS and during SARS. For both pharmaceutical stocks and DOW JONES, we got a p-value of 0.40 and 0.44, which means we fail to reject the null hypothesis; our null hypothesis states that before and during SARS, the average price remained the same, indicating that the percent change remained relatively the same between the two time periods.


```python
# perform t-test on pharma stocks for year before compared to year during covid
tResPharm, pResPharm = stats.ttest_ind(abs(sarsPharmStocks['Pct Change']), abs(presarsPharmStocks['Pct Change']), nan_policy = 'omit')
print('Pharmacy Stocks T-Score (significance of difference): ' + str(round(tResPharm, 3)) 
      + '\nPharmacy Stocks P-Value (probabilty this difference is from chance): ' + str(pResPharm))

# perform t-test on DJI stocks for year before compared to year pre covid
tResDJI, pResDJI = stats.ttest_ind(abs(sarsDJI['Pct Change']), abs(presarsDJI['Pct Change']), nan_policy = 'omit')
print('DOW Jones Stocks T-Score (significance of difference): ' + str(round(tResDJI, 3)) 
      + '\nDOW Jones Stocks P-Value (probabilty this difference is from chance): ' + str(pResDJI))
```
